```python
# Speed it up and more!
for df in dataframes.values():
    # Relative Strength Index (RSI)
    delta = df['Adj Close'].diff()
    gain = (delta.where(delta > 0, 0)).fillna(0)
    loss = (-delta.where(delta < 0, 0)).fillna(0)
    avg_gain = gain.rolling(window=14).mean()
    avg_loss = loss.rolling(window=14).mean()
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    # Bollinger Bands
    rolling_mean = df['Adj Close'].rolling(window=20).mean()
    rolling_std = df['Adj Close'].rolling(window=20).std()
    df['Bollinger Upper Band'] = rolling_mean + (rolling_std * 2)
    df['Bollinger Lower Band'] = rolling_mean - (rolling_std * 2)
    
    #  Daily Return
    df['Daily Return'] = df['Adj Close'].pct_change()
    
    # Cumulative Return
    df['Cumulative Return'] = (1 + df['Daily Return']).cumprod()
    
    # 20-Day Volatility
    df['20-Day Volatility'] = df['Daily Return'].rolling(window=20).std()
    
    # Price Volatility
    df['Price Volatility'] = df['Adj Close'].rolling(window=10).std()
    
    # MACD - Moving Average Convergence Divergence
    df['EMA12'] = df['Adj Close'].ewm(span=12, adjust=False).mean()
    df['EMA26'] = df['Adj Close'].ewm(span=26, adjust=False).mean()
    df['MACD'] = df['EMA12'] - df['EMA26']
    df['Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
    df['MACD_Histogram'] = df['MACD'] - df['Signal']
    
    # On Balance Volume (OBV)
    df['Daily Returns'] = df['Adj Close'].pct_change()
    df['Direction'] = [1 if x >= 0 else -1 for x in df['Daily Returns']]
    df['Direction'][0] = 0
    df['OBV'] = (df['Volume'] * df['Direction']).cumsum()
    
    # Price Rate of Change
    df['ROC'] = df['Close'].pct_change(periods=12)
```
